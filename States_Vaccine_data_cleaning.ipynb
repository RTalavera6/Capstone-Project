Create a function that will sort the states into their High education and Low Education Categories.
Apply function to Dataframe.

```python
def state_ed_cat(state):
    if state in ["MA","MD","CT","VT","CO","VA","NJ","NH","MN","WA"]:
        return "High"
    elif state in ["AL","AK","LA","MS","WV","NM","SC","NV","OK","KY"]:
        return "Low"
    else: 
        return "Middle"

Vaccine_Daily["State_Ed"] = Vaccine_Daily["Location"].apply(state_ed_cat)
```

Create a dictionary to pass the amount of fully vaccinated into a function to get the proportion of vaccinated in each state

```python
#Vaccine_Daily.groupby("Location").agg({"Administered":"max"})
Full_Vaccinations = Vaccine_Daily.groupby("Location").agg({"Series_Complete_Yes":"max"}).to_dict(orient = "dict")
Full_Vaccinations = Full_Vaccinations["Series_Complete_Yes"]
Full_Vaccinations.get("AL")
```

Create a Dict of Populations that will be passed into a function to get the proportion of citizens of each state that have been fully vaccinated.

```python
Populations = {"AK": 3011524,
"AL": 5024279,
"CO": 5773714,
"CT": 3605944,
"KY": 4505836,
"LA":4657757,
"MA":7029917,
"MD":6177224,
"MN":5706494,
"MS":6154913,
"NH":1377529,
"NJ":9288994,
"NM":2117522,
"NV":3104614,
"OK":3959353,
"SC":5118425,
"VA":8631393,
"VT":643077,
"WA":7705281,
"WV":1793716}


```

Create Function to find proportion of fully Vaccinated in each state

Once the proportions are found, put them back into the dataframe and then the data is ready for testing.

```python
def state_proportion_finder (Population, Vaccinated):
    Percent_Vacc = {}
    for key in Population:
        Proportion = Vaccinated.get(key)/Population.get(key)
        Percent_Vacc.update({key : Proportion})
    
    return Percent_Vacc
    

State_Vax_Proportions = state_proportion_finder(Populations,Full_Vaccinations)
State_Vax_Proportions

Proportion_Vax_DF= pd.DataFrame.from_dict(State_Vax_Proportions, orient = 'index' ,columns = ['Proportion Vaccinated'])
Proportion_Vax_DF["Location"] = Proportion_Vax_DF.index
Proportion_Vax_DF

Vaccine_Daily = pd.merge(Vaccine_Daily,Proportion_Vax_DF,how = "left", on = "Location")
Vaccine_Daily

```
