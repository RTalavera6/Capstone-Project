```python
import os
import pandas as pd
import datetime
```

```python
pwd = os.getcwd()
```

Import CSV file and save it as a dataframe

```python
filepath = pwd +  "\COVID-19_Vaccinations_in_the_United_States_Jurisdiction.csv"
Vaccine_Daily = pd.read_csv(filepath)
Vaccine_Daily
```

Change Dataframe to only contain the states that will be compared 

```python


Comp_States = ["MA","MD","CT","VT","CO","VA","NJ","NH","MN","WA","AL","AK","LA","MS","WV","NM","SC","NV","OK","KY"]

Vaccine_Daily = Vaccine_Daily[Vaccine_Daily["Location"].isin(Comp_States)]
Vaccine_Daily
```

Seperating dates to better sort data

```python
Vaccine_Daily[["Month","Day","Year"]] = Vaccine_Daily["Date"].str.split("/",expand = True)
Vaccine_Daily
```

Sort Values by Location and then by year... I do it this way because the Date format was causing issues with the changes in the year which was messing with the Data frame presentation and ultimately the visuals themselves.


```python
Vaccine_Daily_sorted = Vaccine_Daily.sort_values(["Location", "Year", "Day","Month"], ascending= True)
Vaccine_Daily_sorted
```
